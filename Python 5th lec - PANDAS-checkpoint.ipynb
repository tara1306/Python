#  Slicing of DataFrames

```python
#Extract records from 30 to 35 rows from Mg to Ca column

df.loc[30:35,'Mg':'Ca']
#loc is location 
#loc allows to put name of columns
#last column/index is included
```

```python
df.iloc[30:35,3:5]  
#i stands for integer location
#iloc doesn't allow to put names of column instead put the index of the column
#displays one less column & row than the given range (last column&row/index is excluded)
```

```python
#Extract only 5th and 18th row from Mg and K

df.loc[[5,18],['Mg','K']]
```

# Pandas Filtering and Groupby

```python
#Extract records where gender = female from Salary.csv file
```

```python
data['gender'] == 'Female'
```

```python
data[data['gender'] == 'Female']
#Masking
```

```python
#Gender is female and salary is greater than 150000

data['gender'] == 'Female'
data['salary'] > 150000

data[(data['gender'] == 'Female') & (data['salary'] > 150000)]
#for and keywor use &
#for or keyword use |
```

```python
data[(data['gender'] == 'Female') | (data['salary'] > 150000)]
```

```python
data[data['rank'] == 'Prof']
```

```python
data['salary'].mean()
```

```python
data[data['rank'] == 'Prof']['salary'].mean()
```

```python
data[(data['rank'] == 'Prof') & (data['gender'] == 'Female')]['salary'].mean()
```

```python
#Groupby aggregate method
```

```python
data.groupby('rank')
```

```python
data.groupby('rank')['salary'].mean()
```

```python
data.groupby('rank')['salary'].agg({'mean','median','min','max'})
```
